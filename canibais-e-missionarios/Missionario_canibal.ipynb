```python
import math

estadoInicial=[3,3,0,0,0]
# [Missionários esq., canibais na esq., missionários dir., canibais dir., posição da canoa (0=esq, 1=dir)]

operadores = [(1,0), (1,1), (2,0), (0,2)] #possiveis operações [(num. missionários, num. canibais)]
borda = []
visitados = [] #vai armazenar os nodos visitados

def deslocarCanoa(estadoAtual, numMissionarios=0, numCanibais=0): #direção = 0 ou 1
  
    if numMissionarios+numCanibais >2: #valida o num de elementos no barco
        return

#     print("estado recebido ->",estadoAtual,numMissionarios,numCanibais)  
#barco do lado esquerdo = 0, lado direito = 1
    if estadoAtual[-1]==0: #a posição -1 é a última da lista, e a última da lista é a posição do barco)
      missionariosOrigem = 0 
      canibaisOrigem = 1
      missionariosDestino = 2
      canibaisDestino = 3
        
    else: #agora a posição 2 e 3 serão as novas origens, pois são as referentes ao lado DIREITO, e o destino será o lado esq
      missionariosOrigem = 2
      canibaisOrigem = 3
      missionariosDestino = 0
      canibaisDestino = 1
        
#se os elementos da origem são maiores que zero, se não forem return
    if estadoAtual[missionariosOrigem]== 0 and estadoAtual[canibaisOrigem]==0:
        return

#att o lado da canoa 
    estadoAtual[-1] = 1 - estadoAtual[-1] #se estiver no lado direito (1), será 1-1, se estiver no esq. (0) será 1-0

    for i in range(min(numMissionarios, estadoAtual[missionariosOrigem])): #aha o menor número entre o num de missionários q quer passar e os missionarios disponíveis
      estadoAtual[missionariosOrigem]-=1 #diminui na orirem e aumenta no destino
      estadoAtual[missionariosDestino]+=1

    for i in range(min(numCanibais, estadoAtual[canibaisOrigem])):
        estadoAtual[canibaisOrigem]-=1 #diminui na orirem e aumenta no destino
        estadoAtual[canibaisDestino]+=1

    return estadoAtual

# calcular os estados alcançaveis Validos a partir de um estado para cada operador, aplicando REGRAS
def sucessores(estadoAtual): 
  arraySucessores = [] #armazenar os sucessores

  for (i,j) in operadores:
    sucessor = deslocarCanoa(estadoAtual[:], i, j) #copia do array, i seria os missionários e j os canibais
    
    if sucessor == None: #se estiver vazio
        continue 
    
    if(sucessor[0] < sucessor[1] and sucessor[0] > 0) or (sucessor[2] < sucessor[3] and sucessor[2] > 0): 
        continue #missionarios menor que num de canibais e maior que 0, no destino OU origem.
    
    if sucessor in visitados: 
        continue; #verificação de ESTADOS REPETIDOS
    arraySucessores.append(sucessor)
  
  return arraySucessores

def calculaFx(estadoAtual):#custo de chegada ao no atual G(n) + h(n)
    esquerda = (estadoAtual[0], estadoAtual[1])
    direita = (estadoAtual[2], estadoAtual[3])
    
    custoEncontrado = math.dist(esquerda,direita) # Calculate Euclidean distance

    heuristica = estadoAtual[0] + estadoAtual[1] #canibais e missionários lado esq é a heuristica
    
    fx = custoEncontrado + heuristica 
    
    return fx

def obtemAdjacenteNaoVisitado(estadoParaAnalisar): #utiliza a func calculaFx
    salvaCustos = []  
    adjacente = sucessores(estadoParaAnalisar)# chama o sucessores, passa pelas regras.. 

    if len(adjacente) > 0:
        for x in adjacente:
         salvaCustos.append(calculaFx(x))
    
        menorValor = min(salvaCustos)
        return adjacente[salvaCustos.index(menorValor)]

    else:
        return -1 #sem estado adjacente

def testeFinal(estadoFinal):
    if estadoFinal[2]== 3 and estadoFinal[3] == 3: # [0,0,3,3,1]
        return True
    else:
        return False

def busca(estadoInicial):
    borda.append(estadoInicial)
    while len(borda) != 0:
        estadoParaAnalisar = borda[len(borda)-1]
    
        if testeFinal(estadoParaAnalisar): break
        retornoAdjacente = obtemAdjacenteNaoVisitado(estadoParaAnalisar)
        if retornoAdjacente == -1:#se retorna -1, não existe então tira o elemento
          borda.pop()
        else:#se não retorna -1, adiciona aos visitados e adiciona a borda
          visitados.append(retornoAdjacente)
          borda.append(retornoAdjacente)
    else:
        print("Caminho não encontrado")
    return borda #se estado atual for a borda, será listado o caminho do estado inicial ate o estado final
solucao = busca(estadoInicial)
print("Caminho percorrido:", solucao)
```
